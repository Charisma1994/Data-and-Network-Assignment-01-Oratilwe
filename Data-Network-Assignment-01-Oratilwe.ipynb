# Introduction

The data below is an educational data set which is collected from learning management system (LMS) called Kalboard 360, describe the performance of students from various Grades and Countries in a semester and various factors which play a role in their performance.

The data is collected using a learner activity tracker tool, which called experience API (xAPI). 
The xAPI is a component of the training and learning architecture (TLA) that enables to monitor learning progress and learnerâ€™s actions like reading an article or watching a training video. 
The API helps the learning activity providers to determine the learner, activity and those objects that describe a learning experience. The dataset below consists of 480 student records and 16 features. The features are classified into three categories: (1) Demographic features such as gender and nationality. (2) Academic background features such as educational stage, grade Level and section. (3) Behavioral features such as raised hand on class, opening resources, answering survey by parents, and school satisfaction.

I'm trying to determine what among the factors listed above actually affects the students performance is it the relationship they have with their parents or not?, and  also determine some measures which one can take in order to diagnose the problem and find alternative measures to boasting the student's education performance.

# Type of method: 
For this problem I used the Logistic Regression since my data was categorical to determine the association between the student's relationship their parents either mother or father and their performance.

```python
%matplotlib inline
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
os.chdir("/home/oratilwe/Desktop/MAXIMA/DataScience/Data-Networks-Assignment-01-Oratilwe")
```

```python
df = pd.read_csv("xAPI-Edu-Data.csv")
```

```python
formula = 'df.Relation ~ df.raisedhands + df.VisITedResources + df.AnnouncementsView + df.Discussion + df.StudentAbsenceDays'
```

```python
mod = smf.glm(formula = formula, data = df, family = sm.families.Binomial())
```

```python
results = mod.fit()
```

```python
print(results.summary())
```

# Analysis

Based on the GLM regression results we can see that there is an association between the relation(Mum and Father) and student abscence days, thus one can say that the one factor which contribute to the student's poor performance is the number of days they are absent from.
